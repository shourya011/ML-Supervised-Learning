```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1,2])
y_train = np.array([300,500])
```

```python
def computer_cost(w,b,x,y):
    m = x.shape[0]
    cost = 0
    for i in range(m):
        f_wb = w*x[i] + b
        cost_sum = (f_wb - y[i]) ** 2
        cost = cost + cost_sum
    totol_sum = (1/(2*m)) * cost

    return totol_sum
```

```python
w = 100
b = 100
```

```python
computer_cost(w,b,x_train,y_train)
```
