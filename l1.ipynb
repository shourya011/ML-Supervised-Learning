```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0,2.0]);
y_train= np.array([300,500]);
print(f"the value is x_train : {x_train}")
print(f"the value is y_train: {y_train}")
```

```python
m = x_train.shape[0];
print(f"this Size is: {m}")
l = len(x_train);
print(f"the length is : {l}")
```

```python
i = 1
x_i = x_train[1];
y_i = y_train[1];
print(f"x^{i} and y^{i} : {x_i},{y_i}")
```

```python
plt.scatter(x_train,y_train,marker='X',c='r')
plt.title("House price prediction")
plt.xlabel("Size of house in 1000 sqft")
plt.ylabel("price of house in $1000")
plt.show()
```

```python
w = 150
b = 175
print(f"weight : {w}")
print(f"bais : {b}")
```

```python
def computer_output(x,w,b):
    m = x.shape[0]
    f_wb = np.zeros(m);
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
temp = computer_output(x_train,w,b)
plt.plot(x_train,temp,c='b',label="prediction")
plt.scatter(x_train,y_train,c='r',marker='X',label="actual")  
plt.title("House price prediction")
plt.xlabel("Size of house in 1000 sqft")
plt.ylabel("price of house in $1000")
plt.legend()
plt.show()
```
